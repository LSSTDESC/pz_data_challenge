```python
import yaml
import matplotlib.pyplot as plt
import numpy as np
import pandas
from pz_data_challenge.evaluation import build_summary_data_dict, make_strip_plot
```

```python
RESULTS_DIR = '../results'
SUBMISSIONS = ['easy_forest', 'rail_knn_test', 'rail_bpz_test', 'bpz_31temps']

Y_LABEL_STRINGS = [
    'taskset_1_1yr_cardinal',
    'taskset_1_1yr_flagship',
    'taskset_1_10yr_cardinal',
    'taskset_1_10yr_flagship',
    'taskset_2_1yr_cardinal',
    'taskset_2_1yr_flagship',
    'taskset_2_10yr_cardinal',
    'taskset_2_10yr_flagship',
]
```

```python
data_dict = build_summary_data_dict(RESULTS_DIR, SUBMISSIONS, "point")
```

```python
dd_outlier = {}
dd_CvM = {}
dd_ks = {}
dd_ksamp = {}

for sub_ in SUBMISSIONS:
    task1_mask = data_dict[sub_]['task'] == 1 
    run_ = 4*(data_dict[sub_]['taskset']-1) + 2*(data_dict[sub_]['scenario']-1) + (data_dict[sub_]['sim']-1)
    dd_outlier[sub_] = (data_dict[sub_]['outlier'][task1_mask], run_[task1_mask])
    dd_CvM[sub_] = (data_dict[sub_]['CvM'][task1_mask], run_[task1_mask])
    dd_ks[sub_] = (data_dict[sub_]['ks'][task1_mask], run_[task1_mask])
    dd_ksamp[sub_] = (data_dict[sub_]['ksamp'][task1_mask], run_[task1_mask])    
```

```python
fig_CvM = make_strip_plot(
    dd_CvM,
    Y_LABEL_STRINGS,
    r"PIT Q-Q CvM Metric",
    [0, 2000],
    [[0, 250], [0, 500], [0, 1000]],
)
```

```python
fig_outlier = make_strip_plot(
    dd_outlier,
    Y_LABEL_STRINGS,
    r"PIT Q-Q Outliers",
    [0, 1],
    [[0, 0.2], [0, 0.4], [0, 0.6]],
)
```

```python
fig_ks = make_strip_plot(
    dd_ks,
    Y_LABEL_STRINGS,    
    r"PIT Q-Q KS",
    [0, 1],
    [[0, 0.2], [0, 0.4], [0, 0.6]],
)
```

```python
fig_ksamp = make_strip_plot(
    dd_ksamp,
    Y_LABEL_STRINGS,        
    r"PIT Q-Q ksamp",
    [0, 1e4],
    [[0, 1e3], [0, 2e3], [0, 4e3]],
)
```

```python
mask_1_1 = np.bitwise_and(data_dict['easy_forest']['taskset'] == 1, data_dict['easy_forest']['task'] == 1)
```

```python
mask_1_1_1_yr = np.bitwise_and(mask_1_1, data_dict['easy_forest']['scenario'] == 1)
mask_1_1_10_yr = np.bitwise_and(mask_1_1, data_dict['easy_forest']['scenario'] == 2)
```

```python
out_data = {}
out_data['easy_forest'] = {f"{k}_1yr":np.mean(data_dict['easy_forest'][k][mask_1_1_1_yr]) for k in ['mean', 'std']}
out_data['easy_forest'].update(**{f"{k}_10yr":np.mean(data_dict['easy_forest'][k][mask_1_1_10_yr]) for k in ['mean', 'std']})
```
