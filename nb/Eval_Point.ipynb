```python
import yaml
import matplotlib.pyplot as plt
import numpy as np
import os
import tables_io
import pandas
```

```python
RESULTS_DIR = '../results'
SUBMISSIONS = ['easy_forest', 'rail_knn_test', 'rail_bpz_test', 'bpz_31temps']
```

```python
data_dict = {}
for sub_ in SUBMISSIONS:
    submission_file = os.path.join(RESULTS_DIR, f"summary_{sub_}_point.parq")
    data_dict[sub_] = tables_io.read(submission_file)
```

```python
for sub_ in SUBMISSIONS:
    taskset1_mask = data_dict[sub_]['taskset'] == 1    
    plt.hist(data_dict[sub_]['abs_outlier_rate'][taskset1_mask], bins=np.linspace(0, 0.1, 101), label=sub_, alpha=0.5)
_ = plt.legend()
```

```python
for sub_ in SUBMISSIONS:
    taskset2_mask = data_dict[sub_]['taskset'] == 2    
    plt.hist(data_dict[sub_]['abs_outlier_rate'][taskset2_mask], bins=np.linspace(0, 1.0, 101), label=sub_, alpha=0.5)
_ = plt.legend()
```

```python
for sub_ in SUBMISSIONS:
    task1_10yr_mask = np.bitwise_and(data_dict[sub_]['task'] == 1, data_dict[sub_]['scenario'] == 2)
    plt.scatter(data_dict[sub_]['taskset'][task1_10yr_mask], data_dict[sub_]['abs_outlier_rate'][task1_10yr_mask], label=sub_)
_ = plt.legend()
```

```python
for sub_ in SUBMISSIONS:
    task1_1yr_mask = np.bitwise_and(data_dict[sub_]['task'] == 1, data_dict[sub_]['scenario'] == 1)
    plt.scatter(data_dict[sub_]['taskset'][task1_1yr_mask], data_dict[sub_]['abs_outlier_rate'][task1_1yr_mask], label=sub_)
_ = plt.legend()
```

```python
for sub_ in SUBMISSIONS:
    run_ = 16*(data_dict[sub_]['taskset']-1) + 8*(data_dict[sub_]['scenario']-1) + 4*(data_dict[sub_]['sim']-1) +  (data_dict[sub_]['task']-1)
    plt.scatter(run_, data_dict[sub_]['abs_outlier_rate'], label=sub_)
_ = plt.legend()
```

```python
for sub_ in SUBMISSIONS:
    run_ = 16*(data_dict[sub_]['taskset']-1) + 8*(data_dict[sub_]['scenario']-1) + 4*(data_dict[sub_]['sim']-1) +  (data_dict[sub_]['task']-1)
    plt.scatter(run_, data_dict[sub_]['mean'], label=sub_)
_ = plt.legend()
```

```python
for sub_ in SUBMISSIONS:
    task1_mask = data_dict[sub_]['task'] == 1 
    run_ = 4*(data_dict[sub_]['taskset']-1) + 2*(data_dict[sub_]['scenario']-1) + (data_dict[sub_]['sim']-1)
    plt.scatter(run_[task1_mask], data_dict[sub_]['abs_outlier_rate'][task1_mask], label=sub_)
_ = plt.legend()
```

```python
for sub_ in SUBMISSIONS:
    task1_mask = data_dict[sub_]['task'] == 1 
    run_ = 4*(data_dict[sub_]['taskset']-1) + 2*(data_dict[sub_]['scenario']-1) + (data_dict[sub_]['sim']-1)
    plt.scatter(run_[task1_mask], data_dict[sub_]['std'][task1_mask], label=sub_)
_ = plt.legend()
```
