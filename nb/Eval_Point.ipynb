```python
import yaml
import matplotlib.pyplot as plt
import numpy as np
import os
import tables_io
import pandas
```

```python
RESULTS_DIR = '../results'
SUBMISSIONS = ['easy_forest', 'rail_knn_test', 'rail_bpz_test', 'bpz_31temps']

Y_LABEL_STRINGS = [
    'taskset_1_1yr_cardinal',
    'taskset_1_1yr_flagship',
    'taskset_1_10yr_cardinal',
    'taskset_1_10yr_flagship',
    'taskset_2_1yr_cardinal',
    'taskset_2_1yr_flagship',
    'taskset_2_10yr_cardinal',
    'taskset_2_10yr_flagship',
]
```

```python
data_dict = {}
for sub_ in SUBMISSIONS:
    submission_file = os.path.join(RESULTS_DIR, f"summary_{sub_}_point.parq")
    data_dict[sub_] = tables_io.read(submission_file)
```

```python
def plot_v_run_multi(
    data,
    metric_label,
    metric_limits,
    metric_ranges,
):
    fig = plt.Figure()
    for key, val in data.items():
        plt.scatter(val[0], val[1], label=key)
    _ = plt.yticks(np.linspace(1, 29, 8), Y_LABEL_STRINGS)
    _ = plt.xlabel(metric_label)
    _ = plt.ylim(-0.5, 31.5)
    _ = plt.xlim(metric_limits)
    for metric_range in metric_ranges:        
        _ = plt.fill_between(metric_range, [-0.5, -0.5], [31.5, 31.5], color='gray', alpha=0.1)
    _ = plt.legend()
    return fig

def plot_v_run(
    data,
    metric_label,
    metric_limits,
    metric_ranges,
):
    fig = plt.Figure()
    for key, val in data.items():
        plt.scatter(val[0], val[1], label=key)
    _ = plt.yticks(np.linspace(0, 7, 8), Y_LABEL_STRINGS)
    _ = plt.xlabel(metric_label)
    _ = plt.ylim(-0.5, 7.5)
    _ = plt.xlim(metric_limits)
    for metric_range in metric_ranges:        
        _ = plt.fill_between(metric_range, [-0.5, -0.5], [7.5, 7.5], color='gray', alpha=0.1)
    _ = plt.legend()
    return fig
```

```python
dd_outliers = {}
dd_mean = {}
dd_rms = {}

for sub_ in SUBMISSIONS:
    task1_mask = data_dict[sub_]['task'] == 1 
    run_ = 4*(data_dict[sub_]['taskset']-1) + 2*(data_dict[sub_]['scenario']-1) + (data_dict[sub_]['sim']-1)
    dd_mean[sub_] = (data_dict[sub_]['mean'][task1_mask], run_[task1_mask])
    dd_rms[sub_] = (data_dict[sub_]['std'][task1_mask], run_[task1_mask])
    dd_outliers[sub_] = (data_dict[sub_]['abs_outlier_rate'][task1_mask], run_[task1_mask])
```

```python
fig_mean = plot_v_run(
    dd_mean,
    r"Mean $\frac{|z_{\rm est} - z_{\rm ref}|}{1 + z_{\rm ref}}$",
    [-0.2, 0.2],
    [[-0.01, 0.01], [-0.025, 0.025], [-0.05, 0.05]],
)
```

```python
fig_rms = plot_v_run(
    dd_rms,
    r"RMS $\frac{|z_{\rm est} - z_{\rm ref}|}{1 + z_{\rm ref}}$",
    [0, 0.5],
    [[0.0, 0.025], [0.0, 0.050], [0.0, 0.10]],
)
```

```python
fig_outliers = plot_v_run(
    dd_outliers,
    r"Outlier rate $\frac{|z_{\rm est} - z_{\rm ref}|}{1 + z_{\rm ref}} > 0.15$",
    [0, 0.5],
    [[0.0, 0.025], [0.0, 0.1], [0.0, 0.2]],
)
```

```python
dd_outliers_all = {}
dd_mean_all = {}
dd_rms_all = {}

for sub_ in SUBMISSIONS:
    run_ = 16*(data_dict[sub_]['taskset']-1) + 8*(data_dict[sub_]['scenario']-1) + 4*(data_dict[sub_]['sim']-1) +  (data_dict[sub_]['task']-1)
    dd_mean_all[sub_] = (data_dict[sub_]['mean'], run_)
    dd_rms_all[sub_] = (data_dict[sub_]['std'], run_)
    dd_outliers_all[sub_] = (data_dict[sub_]['abs_outlier_rate'], run_)
```

```python
fig_mean = plot_v_run_multi(
    dd_mean_all,
    r"Mean $\frac{|z_{\rm est} - z_{\rm ref}|}{1 + z_{\rm ref}}$",
    [-0.2, 0.2],
    [[-0.025, 0.025], [-0.1, 0.1]],
)
```

```python
fig_rms = plot_v_run_multi(
    dd_rms_all,
    r"RMS $\frac{|z_{\rm est} - z_{\rm ref}|}{1 + z_{\rm ref}}$",
    [0, 0.5],
    [[0.0, 0.025], [0.0, 0.050]],
)
```

```python
fig_outliers = plot_v_run_multi(
    dd_outliers_all,
    r"Outlier rate $\frac{|z_{\rm est} - z_{\rm ref}|}{1 + z_{\rm ref}} > 0.15$",
    [0, 0.5],
    [[0.0, 0.025], [0.0, 0.050]],
)
```
