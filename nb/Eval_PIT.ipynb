```python
import yaml
import matplotlib.pyplot as plt
import numpy as np
import os
import tables_io
import pandas
```

```python
RESULTS_DIR = '../results'
SUBMISSIONS = ['easy_forest', 'rail_knn_test', 'rail_bpz_test', 'bpz_31temps']
```

```python
data_dict = {}
for sub_ in SUBMISSIONS:
    submission_file = os.path.join(RESULTS_DIR, f"summary_{sub_}_pit_prob.parq")
    data_dict[sub_] = tables_io.read(submission_file)
```

```python
fig = plt.Figure((8, 8))
for i_sub, sub_ in enumerate(SUBMISSIONS):
    for i_row, row_ in data_dict[sub_].iterrows():        
        if row_['task'] != 1:
            continue
        if row_['scenario'] != 1:
            continue
        if row_['sim'] != 1:
            continue    
        cdf = np.cumsum(row_['y_vals'].clip(0, 5))
        lines = ['', '-', 'dashed']
        colors = ['blue', 'orange', 'green', 'red']                   
        plt.plot(row_['x_vals'], cdf/cdf[-1], ls=lines[row_['taskset']], c=colors[i_sub], label=f"{sub_} {row_['taskset']}")
_ = plt.plot([0,1],[0,1])
_ = plt.xlim(0, 1)
_ = plt.ylim(0, 1)
_ = plt.legend()
```
