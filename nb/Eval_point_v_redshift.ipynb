```python
import yaml
import matplotlib.pyplot as plt
import numpy as np
import os
import tables_io
import pandas
```

```python
RESULTS_DIR = '../results'
SUBMISSIONS = ['easy_forest', 'rail_knn_test', 'rail_bpz_test', 'bpz_31temps']
```

```python
data_dict = {}
for sub_ in SUBMISSIONS:
    submission_file = os.path.join(RESULTS_DIR, f"summary_{sub_}_point_v_redshift.parq")
    data_dict[sub_] = tables_io.read(submission_file)
```

```python
for sub_ in SUBMISSIONS:
    task1_mask = np.bitwise_and(data_dict[sub_]['task'] == 1, data_dict[sub_]['taskset'] == 1)
    for i_row, row_ in data_dict[sub_][task1_mask].iterrows():        
        if row_['taskset'] == 2:
            continue
        if row_['task'] != 1:
            continue
        if row_['scenario'] == 1:
            continue
        if row_['sim'] == 1:
            continue            
        plt.plot(row_['z_mean'], row_['biweight_mean'], label=sub_)
_ = plt.legend()
```
