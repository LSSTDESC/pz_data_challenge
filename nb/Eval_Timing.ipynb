```python
import yaml
import matplotlib.pyplot as plt
import numpy as np
import os
import tables_io
import pandas
```

```python
RESULTS_DIR = '../results'
SUBMISSIONS = ['easy_forest', 'rail_knn_test', 'rail_bpz_test', 'bpz_31temps']
```

```python
data_dict = {}
for sub_ in SUBMISSIONS:
    submission_file = os.path.join(RESULTS_DIR, f"summary_{sub_}_timing.parq")
    data_dict[sub_] = tables_io.read(submission_file)
```

```python
for sub_ in SUBMISSIONS:
    task2_mask = data_dict[sub_]['task'] == 2    
    plt.hist(data_dict[sub_]['time'][task2_mask], bins=np.linspace(0, 25, 101), label=sub_, alpha=0.5)
_ = plt.legend()
```

```python
for sub_ in SUBMISSIONS:
    task3_mask = data_dict[sub_]['task'] == 3    
    plt.hist(data_dict[sub_]['time'][task3_mask], bins=np.linspace(0, 50, 101), label=sub_, alpha=0.5)
_ = plt.legend()
```
